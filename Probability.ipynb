```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/car_data_250.csv"
df = pd.read_csv(file_path)
df
```

```python
# Assuming 'df' is your DataFrame
df.describe(include='all')
# OR
df.info()
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(df['Price'], bins = 20, kde = 1, color = 'blue')
plt.title('Distribution of Car Prices', fontsize = 14)
plt.xlabel('Price{$}', fontsize = 12)
plt.ylabel('Frequency', fontsize = 14)
plt.show()
```

```python
# Most Common Price Range
most_common_price_range = (df['Price'].min(), df['Price'].max())
print(f"Most Common Price Range: {most_common_price_range}")

# 2. Scatterplot Exploration: Horsepower vs. Mileage
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Horsepower'], y=df['Mileage'], hue=df['Fuel Type'], palette='viridis', alpha=0.8)
plt.xlabel('Horsepower (HP)', fontsize=12)
plt.ylabel('Mileage (MPG)', fontsize=12)
plt.title('Horsepower vs Mileage', fontsize=14)
sns.regplot(x=df['Horsepower'], y=df['Mileage'], scatter=False, color='red', ci=None)
plt.legend(title="Fuel Type", bbox_to_anchor=(1, 1), loc='upper left')
plt.show()

# Correlation between Horsepower and Mileage
correlation = df[['Horsepower', 'Mileage']].corr().iloc[0,1]
print(f"Correlation between Horsepower and Mileage: {correlation}")
```
